```python
import requests
from bs4 import BeautifulSoup
```

```python
link = "https://jobs.jnj.com/en/jobs/?search=Data+scientist&country=Switzerland&pagesize=20#results"
```

```python
def parse_link(url):
    try:
        # Make a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract and print the title of the page
            title = soup.title.text
            print(f'Title: {title}')

            # Extract and print other information as needed
            # Example: Extracting all the links on the page
            links = soup.find_all('a')
            print('\nLinks on the page:')
            for link in links:
                print(link.get('href'))

        else:
            print(f'Error: Unable to fetch content. Status code: {response.status_code}')

    except Exception as e:
        print(f'Error: {str(e)}')

# Example usage:
parse_link(link)
```
